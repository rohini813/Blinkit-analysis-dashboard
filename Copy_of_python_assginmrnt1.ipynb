```python
#Pass / Fail Analyzer
```

```python
marks=[45, 78, 90, 33, 60]
```

```python
pass_count=0
fail_count=0
```

```python
for m in marks:
  if m>=50:
    pass_count+=1
  else:
    fail_count+=1

print("TOtal pass student:",pass_count)
print("Total fail student:",fail_count)
```

```python
#Simple Data Cleaner
```

```python
names = [" Alice ", "bob", " CHARLIE "]
```

```python
cleaned_names=[]
```

```python
for name in names:
  clean_name= name.strip().lower()
  cleaned_names.append(clean_name)
print("Cleaned Names:",cleaned_names)
```

```python
#Message Length Analyzer
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
```

```python
for msg in messages:
    length=len(msg)
    print("Message:",msg)
    print("Length:",length)

    if length > 10:
      print("This message is longer than 10 characters")

    print()
```
